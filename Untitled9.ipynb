```python
import json
import faiss
import numpy as np
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer

# Load extracted data
with open("extracted_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Initialize Hugging Face Embeddings Model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Convert data into LangChain Document format
documents = [Document(page_content=item["content"], metadata=item["metadata"]) for item in data]

# Split documents into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(documents)

# Convert text into vector embeddings
texts = [doc.page_content for doc in split_docs]
vectors = embedding_model.encode(texts)

# Convert to numpy array for FAISS
vectors = np.array(vectors, dtype=np.float32)

# Store embeddings in FAISS vector database
index = faiss.IndexFlatL2(vectors.shape[1])  # Create FAISS index
index.add(vectors)  # Add vectors to index

# Save FAISS index
faiss.write_index(index, "faiss_index.bin")

# Save document metadata separately
with open("metadata.json", "w", encoding="utf-8") as f:
    json.dump([doc.metadata for doc in split_docs], f, indent=4)

print("✅ Embeddings generated and stored in FAISS.")
```

```python
from flask import Flask, request, jsonify
import faiss
import numpy as np
import json
from sentence_transformers import SentenceTransformer

# Load FAISS index
index = faiss.read_index("faiss_index.bin")

# Load metadata (course details)
with open("metadata.json", "r", encoding="utf-8") as f:
    metadata = json.load(f)

# Load Sentence Transformer model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Initialize Flask app
app = Flask(__name__)

@app.route("/search", methods=["POST"])
def search():
    """
    Search for relevant courses based on user query.
    """
    try:
        # Get query from request
        data = request.json
        query = data.get("query", "")

        if not query:
            return jsonify({"error": "Query cannot be empty"}), 400

        # Convert query to embedding
        query_vector = embedding_model.encode([query])
        query_vector = np.array(query_vector, dtype=np.float32)

        # Search FAISS index
        k = 3  # Number of results to return
        distances, indices = index.search(query_vector, k)

        # Fetch results
        results = []
        for idx in indices[0]:
            if idx < len(metadata):
                results.append(metadata[idx])

        return jsonify({"query": query, "results": results})

    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)
```
